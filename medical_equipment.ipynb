```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Medical-Equipment-Suppliers.csv')
```

```python
df.describe()
```

```python
# ...existing code...
# New EDA cells
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

print("Shape:", df.shape)
print("\nInfo:")
df.info()

print("\nDescribe (all):")
display(df.describe(include='all').T)

# Missing values
missing = df.isna().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.concat([missing, missing_pct], axis=1, keys=['missing', 'pct_missing']).sort_values('pct_missing', ascending=False)
display(missing_df.head(20))

# Duplicates
print("\nDuplicate rows:", df.duplicated().sum())

# Unique counts
display(df.nunique().sort_values(ascending=False))

# Categorical summaries
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
for c in cat_cols:
    print(f"\n== {c} (top 10) ==")
    print(df[c].value_counts(dropna=False).head(10))

# Numeric distributions and correlation
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if num_cols:
    df[num_cols].hist(bins=20, figsize=(12, max(4, len(num_cols)*1.5)))
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
    plt.title("Numeric Correlation")
    plt.show()
else:
    print("No numeric columns to plot.")

# Example: top suppliers if such column exists
for candidate in ['Supplier', 'Vendor', 'Company', 'Supplier Name']:
    if candidate in df.columns:
        print(f"\nTop 10 by count ({candidate}):")
        print(df[candidate].value_counts().head(10))

# Sample rows with any missing data
display(df[df.isna().any(axis=1)].head())

# Save missing-value report
missing_df.to_csv('eda_missing_report.csv')
print("Saved missing value report: eda_missing_report.csv")
# ...existing code...
```
